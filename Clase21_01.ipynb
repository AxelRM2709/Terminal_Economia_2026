```python
for i in range(5):
  if i == 2:
    break
  print(i)
print("\n")
for i in range(5):
   if i == 2:
    continue
   print(i)
print("\n")
for i in range(5):
  if i == 2:
    pass
  print(i)

print("\n")
import random
print(random.randint(1,10))
adivinar = random.randrange(1,10)
while(1):
  n = int(input("Adivina el número (1-10): "))
  if n == 0:
    break
  if n > 10:
    print("Tu número no está en el rango (1-10)!")
    continue
  elif n == adivinar:
    print("Felicidades, adivinaste el número")
    break
  else:
    print("Vuelve a intentarlo")

```
