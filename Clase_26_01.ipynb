### Conjuntos

```python
vacio = set()
conjunto = {1,2,3,3} # = {1,2,3} no tiene elementos repetidos
l1 = ["1", "2"]
l1_sin_repetidos = set(l1)
conjunto.add(4) # Agregar a un conjunto
conjunto.remove(3) # Eliminar elementos
conjunto.discard(2) # Eliminar de forma segura
conjunto.pop() # Elimina por posición
c1 = {1,2,3,4}
c2 = {2,4,6}
1 in c1 # Verificar que existe
inter = c1 & c2 # Intersección
c1.intersection(c2) # También intersección
union_c = c1 | c2
c1.union(c2)
diff = c1 - c2 # Diferencia
c1.difference(c2)
c1.issubset(c2) # Ver si es subconjunto
c1.issuperset(c2) # Ver si c1 contiene a c2
c1 ^ c2 # c1 unión c2 menos la intersección (diferencia simetrica)
c1.symmetric_difference(c2)
c1 == c2 # no importa el orden
```

### Diagrama de Venn

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles
A = {1,2,3}
B = {3,4,5}
plt.figure(figsize = (6, 6))
venn = venn2([A,B], ('A', 'B'))
plt.show() # Hasta ahora, solo mostrará las cardinalidades
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles
A = {1,2,3}
B = {3,4,5}
plt.figure(figsize = (6, 6))
venn = venn2([A,B], ('A', 'B'))
venn.get_label_by_id('10').set_text('1','2')
venn.get_label_by_id('01').set_text('4','5')
venn.get_label_by_id('11').set_text('3')
plt.show()
```
